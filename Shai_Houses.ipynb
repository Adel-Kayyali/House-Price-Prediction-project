```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

**Getting the DF**

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.describe()
```

**Handling Null Values**

```python
df.isnull().sum()
```

```python
df.drop_duplicates
df=df.drop(['Alley','Utilities','MSZoning','Exterior1st','Exterior2nd','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','KitchenQual','Functional','SaleType','GarageArea','GarageCars','LotFrontage','Id','MiscFeature','Fence','PoolQC','GarageCond','GarageQual','GarageFinish','GarageYrBlt','GarageType','FireplaceQu','Electrical','BsmtFinType2','BsmtFinType1','BsmtExposure','BsmtCond','BsmtQual','MasVnrArea','MasVnrType'],axis=1)
df.columns
```

```python
df.isna().sum().sum()
```

```python
df.shape
```

**Handling Categorical Features with Dummies**

```python
categ_cols = df.select_dtypes('object').columns
categ_cols = categ_cols.to_list()
categ_cols
```

```python
df=pd.get_dummies(df,columns=['Street',
 'LotShape',
 'LandContour',
 'LotConfig',
 'LandSlope',
 'Neighborhood',
 'Condition1',
 'Condition2',
 'BldgType',
 'HouseStyle',
 'RoofStyle',
 'RoofMatl',
 'ExterQual',
 'ExterCond',
 'Foundation',
 'Heating',
 'HeatingQC',
 'CentralAir',
 'PavedDrive',
 'SaleCondition'], drop_first=True)
```

**Splitting the Data**

```python
X=df.drop(columns='SalePrice', axis=1)
y=df['SalePrice']


X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.14,random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Scaling the DF**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_Scaled = scaler.fit_transform(X_train)
X_train_Scaled
```

```python
X_test_Scaled= scaler.transform(X_test)
X_test_Scaled
```

**Selecting & Training Model**

Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train_Scaled, y_train)
```

```python
from sklearn.metrics import mean_squared_error

predictions = lin_reg.predict(X_test_Scaled)
lin_rmse = mean_squared_error(y_test, predictions, squared=False)
lin_rmse
```

SVM - Linear

```python
from sklearn.svm import SVR

svm_reg = SVR(kernel="linear")
```

```python
svm_reg.fit(X_train_Scaled, y_train)
predictions = svm_reg.predict(X_test_Scaled)
svm_mse = mean_squared_error(y_test, predictions)
svm_rmse = np.sqrt(svm_mse)
svm_rmse
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=140, max_features="auto", random_state=44)
```

```python
rf_reg.fit(X_train_Scaled, y_train)
predictions = rf_reg.predict(X_test_Scaled)
rf_mse = mean_squared_error(y_test, predictions)
rf_rmse = np.sqrt(rf_mse)
rf_rmse
```
